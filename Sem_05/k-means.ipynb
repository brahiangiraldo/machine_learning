# K Means - Generated data

Isotropic dataset module in sklearn [here](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_blobs.html) \
Train-test splits in sklearn [here](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html) \
k-Means module in sklearn [here](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html)

```python
import numpy as np 
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
```

```python
_, axes = plt.subplots(1,2, figsize=(8,4))

X,_ = make_blobs(n_samples = 1000 , random_state = 170 )

# Create train (80%) and test (20%) sets and print all shapes
X_train,X_test = train_test_split(X, test_size=0.2, random_state=1)

print(X_train.shape, X_test.shape)
axes[0].scatter(X[:, 0], X[:,1] , edgecolors='k', cmap=plt.cm.Paired)
axes[1].scatter(X_train[:, 0], X_train[:,1] , edgecolors='r')
axes[1].scatter(X_test[:, 0], X_test[:,1] , edgecolors='k' )
```

```python
kmeans = KMeans(n_clusters=3, random_state=1)

# Train Kmeans
kmeans.fit(X_train)
```

```python
# Plot centroids
centroids = kmeans.cluster_centers_
labels = kmeans.labels_
print(centroids),labels.shape
```

```python
# Create predictions
predictions = kmeans.predict(X_test)
print(predictions.shape) 
```

```python
fig, ax = plt.subplots(figsize=(5,4))
colors=['blue','cyan','magenta']
ax.scatter(X_train[:,0], X_train[:,1], c=labels, cmap=plt.cm.Paired, edgecolors='k')
ax.scatter(centroids[:,0], centroids[:,1], c=colors, s=200, marker='x')
ax.scatter(X_test[:,0], X_test[:,1], c=predictions, cmap=plt.cm.Paired, edgecolors='k')

# plot data and found centroids
```
