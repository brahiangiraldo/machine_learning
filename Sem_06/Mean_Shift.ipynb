# Mean Shift - Generated data

Mean shift documentation [here](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.MeanShift.html)

```python
import numpy as np 
import matplotlib.pyplot as plt
# import pandas as pd 
from sklearn.cluster import MeanShift 
from sklearn.datasets import make_blobs 
from matplotlib import pyplot as plt 
from sklearn.model_selection import train_test_split
```

```python
n_samples = 1000 
random_state = 3 
X, Y = make_blobs(n_samples = n_samples, random_state = random_state)

X_train, X_test = train_test_split(X, test_size = 0.20)

_, axes = plt.subplots(1,2, figsize=(10,5))
axes[0].scatter( X[:,0] , X[:,1] , edgecolors='k', cmap=plt.cm.Paired)
axes[1].scatter( X_train[:,0], X_train[:,1], c = 'b', edgecolors='k')
axes[1].scatter( X_test[:,0], X_test[:,1], c = 'r', edgecolors='k')
```

```python
banwidth = max(abs(X[:,0].min() - X[:,0].max()),abs(X[:,1].min() - X[:,1].max()))*0.3
print(banwidth)

ms = MeanShift( bandwidth= banwidth) 
ms.fit(X_train)


cluster_centers = ms.cluster_centers_ 
```

```python
print(cluster_centers.shape ,'\n', cluster_centers)
```

```python
predictions = ms.predict(X_test)

_, ax = plt.subplots(figsize=(6,4))
ax.scatter( X_test[:,0], X_test[:,1], c = predictions , cmap=plt.cm.Paired, edgecolors='k')
ax.scatter(cluster_centers[:,0], cluster_centers[:,1], c = 'm', s = 250, edgecolors='k', marker='X')
```
