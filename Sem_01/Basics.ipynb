```python
import cv2
import numpy as np

# Cargar la imagen
imagen = cv2.imread( "/Users/brahianmichelgiraldoguisao/Documents/Grupo EBD/machine_learning/Images/Caso_1.bmp" )

# Verificar que la imagen se cargó correctamente
if imagen is None:
    print('No se pudo cargar la imagen')
else:
    # Aplicar un filtro de desenfoque
    imagen = cv2.GaussianBlur(imagen, (5, 5), 0)

    # Convertir la imagen a HSV
    hsv = cv2.cvtColor(imagen, cv2.COLOR_BGR2HSV)

    # Definir rangos de color para las fichas
    # Ajusta estos valores según sea necesario
    rojo_bajo = np.array([0, 70, 50])
    rojo_alto = np.array([10, 255, 255])
    verde_bajo = np.array([36, 50, 50])
    verde_alto = np.array([86, 255, 255])
    azul_bajo = np.array([110, 50, 50])
    azul_alto = np.array([130, 255, 255])

    # Crear máscaras para cada color
    mascara_rojo = cv2.inRange(hsv, rojo_bajo, rojo_alto)
    mascara_verde = cv2.inRange(hsv, verde_bajo, verde_alto)
    mascara_azul = cv2.inRange(hsv, azul_bajo, azul_alto)

    # Encontrar contornos en las máscaras
    contornos_rojo = cv2.findContours(mascara_rojo, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)[0]
    contornos_verde = cv2.findContours(mascara_verde, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)[0]
    contornos_azul = cv2.findContours(mascara_azul, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)[0]

    # Usar un umbral de tamaño para los contornos
    umbral_tamano = 100  # Ajusta este valor según sea necesario
    contornos_rojo = [c for c in contornos_rojo if cv2.contourArea(c) > umbral_tamano]
    contornos_verde = [c for c in contornos_verde if cv2.contourArea(c) > umbral_tamano]
    contornos_azul = [c for c in contornos_azul if cv2.contourArea(c) > umbral_tamano]

    # Imprimir el número de fichas de cada color
    print('Número de fichas rojas:', len(contornos_rojo))
    print('Número de fichas verdes:', len(contornos_verde))
    print('Número de fichas azules:', len(contornos_azul))
```
